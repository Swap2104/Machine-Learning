```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
import pickle 
```

```python
df=pd.read_csv("canada_per_capita_income.csv")
df.head()
```

```python
# %matplotlib inline
plt.xlabel('Year')
plt.ylabel('per_capita_income')
plt.scatter(df.year, df.per_capita_income)
```

```python
reg=linear_model.LinearRegression()
```

```python
reg.fit(df[['year']],df.per_capita_income)
```

```python
reg.predict([[2020]])
```

```python
reg.predict([[2017]])
```

```python
# %matplotlib inline
plt.xlabel('Year')
plt.ylabel('per_capita_income')
plt.scatter(df.year, df.per_capita_income)
plt.plot(df.year, reg.predict(df[['year']]) ,color='red')
```

```python
with open('LR_Exercise_Model','wb') as f:
    pickle.dump(reg,f)
```

```python
with open ('LR_Exercise_Model','rb') as f:
    mp = pickle.load(f)
```
