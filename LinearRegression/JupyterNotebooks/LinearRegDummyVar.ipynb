```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
from sklearn.preprocessing import LabelEncoder
```

```python
df=pd.read_csv('homeprices2.csv')
df
```

```python
dum=pd.get_dummies(df.town)
dum
```

```python
merged=pd.concat([df,dum],axis='columns')
merged
```

```python
data=merged.drop(['town','west windsor'],axis='columns')
data
```

```python
reg=linear_model.LinearRegression()
```

```python
x=data.drop(['price'],axis='columns')
x
```

```python
y=data.price
y
```

```python
reg.fit(x,y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
reg.predict([[2888,0,1]])
```

```python
reg.predict([[2388,0,0]])
```

```python
data
```

```python
reg.score(x,y)
```
